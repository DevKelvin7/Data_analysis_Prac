# I. 'seoulCCTV.csv' 파일 전처리

```
import warnings
warnings.filterwarnings('ignore')
```

> ## 1) 'seoulCCTV.csv' 파일 읽어오기

```
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/seoulCCTV.csv'
CCTV = pd.read_csv(url)

CCTV.head()
```

> ## 2) '기관명'을 '구별'로 변경

* 열(Column) 이름 확인

```
CCTV.columns
```

* 첫번째 열 이름 변경('기관명' -> '구별') 

```
CCTV.rename(columns={'기관명':'구별'}, inplace = True)
```

```
CCTV.head()
```

> ## 3) CCTV 데이터 현황 파악

* '소계' 오름차순 정렬

```
CCTV.sort_values('소계')
```

* '소계' 내림차순 정렬

```
CCTV.sort_values('소계', ascending = False)
```

> ## 4) '최근증가율' 열(Column) 추가

* 최근 3년 CCTV '최근증가율' 열 추가

```
CCTV['최근증가율'] = ((CCTV['2016년'] + CCTV['2015년'] + CCTV['2014년']) / CCTV['2013년도 이전']) * 100
```

```
CCTV.head()
```

* '최근증가율' 열로 내림차순 정렬

```
CCTV.sort_values('최근증가율', ascending = False)
```

# II.  'seoulPopulation.xlsx' 파일 전처리

> ## 1) 'seoulPopulation.xlsx' 파일 읽어오기

```
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/seoulPopulation.xlsx'
SDFP = pd.read_excel(url, 
                     header = 2,
                     usecols = 'B, D, G, J, N')

SDFP.head()
```

> ## 2) 열(Column) 이름 변경

```
SDFP.rename(columns={SDFP.columns[0]:'구별', 
                     SDFP.columns[1]:'인구수', 
                     SDFP.columns[2]:'한국인', 
                     SDFP.columns[3]:'외국인', 
                     SDFP.columns[4]:'고령자'}, inplace=True)
```

```
SDFP.head()
```

> ## 3) NaN 확인

* .info( )

```
SDFP.info()
```

* .value_counts( )
 - '합계' 삭제 필요

```
SDFP['구별'].value_counts()
```

* .unique( )

```
SDFP['구별'].unique()
```

> ## 4) 인구 데이터 현황 파악

* '합계' 행(Row) 삭제

```

SDFP.drop(index = 0, axis = 0, inplace = True)
```

```
SDFP.head()
```

* '합계' 행(Row) 삭제 확인

```
SDFP['구별'].value_counts()
```

> ## 5) 파생변수 추가

* '외국인비율'과 '고령자비율' 열(Column)을 계산 후 추가

```
SDFP['외국인비율'] = SDFP['외국인'] / SDFP['인구수'] * 100
SDFP['고령자비율'] = SDFP['고령자'] / SDFP['인구수'] * 100
```

```
SDFP.head()
```

* '인구수' 내림차순 정렬

```
SDFP.sort_values('인구수', ascending = False)
```

* '외국인' 내림차순 정렬

```
SDFP.sort_values('외국인', ascending = False)
```

* '외국인비율' 내림차순 정렬

```
SDFP.sort_values('외국인비율', ascending = False)
```

* '고령자' 내림차순 정렬

```
SDFP.sort_values('고령자', ascending = False)
```

* '고령자비율' 내림차순 정열

```
SDFP.sort_values('고령자비율', ascending = False)
```

# III. CCTV 데이터와 인구 데이터 합치기

> ## 1) 두 데이터프레임에 공통으로 있는 '구별'로 merge

```
CCTV_SDFP = pd.merge(CCTV, SDFP)
```

```
CCTV_SDFP.head()
```

```
CCTV_SDFP.info()
```

> ## 2) 불필요한 열(Column) 삭제

```
CCTV_SDFP.drop(columns = ['2013년도 이전', '2014년', '2015년','2016년'], inplace = True)
```

```
CCTV_SDFP.head()
```

> ## 3) '구별'을 index로 지정

```
CCTV_SDFP.set_index('구별', inplace = True)
```

```
CCTV_SDFP.head()
```

> ## 4) 상관계수(Correlation Coefficient)

* 범위 : -1 ~ 1(0이면 관계없음)
* '고령자비율' vs. '소계'

```
import numpy as np

print(np.corrcoef(CCTV_SDFP['고령자비율'], CCTV_SDFP['소계']))
```

* '외국인비율' vs. '소계'

```
print(np.corrcoef(CCTV_SDFP['외국인비율'], CCTV_SDFP['소계']))
```

* '인구수'     vs. '소계'

```
print(np.corrcoef(CCTV_SDFP['인구수'], CCTV_SDFP['소계']))
```

> ## 5) CCTV 개수('소계')와 '인구수'의 관계

* '소계'(CCTV 개수) 내림차순 정렬

```
CCTV_SDFP.sort_values('소계', ascending = False).head()
```

* '인구수' 내림차순 정렬

```
CCTV_SDFP.sort_values('인구수', ascending = False).head()
```

# IV. Visualization

> ## 1) 막대(bar) 그래프 - pandas

* '소계'

```

CCTV_SDFP[['소계']].plot(kind = 'barh',
                       width = 0.5,
                       figsize = (10, 7), grid = True)
```

* 정렬 - sort_value( )


```
sorted_df1 = CCTV_SDFP.sort_values('소계', ascending = True)
sorted_df1[['소계']].plot(kind = 'barh', width =0.5, figsize = (10, 7), grid = True)
```

* '인구수' 대비 CCTV비율 계산 후 정렬하여 시각화

```

df2 = CCTV_SDFP
df2['인구수대비 CCTV비율'] = CCTV_SDFP['소계'] / CCTV_SDFP['인구수'] * 100
sorted_df2= df2.sort_values('인구수대비 CCTV비율')
```

```
sorted_df2
sorted_df2[['인구수대비 CCTV비율']].plot(kind = 'barh', width =0.5, figsize = (10, 7), grid = True)
```

> ## 2) 산점도(scatter) - matplotlib

* '소계' ~ '인구수'

```
plt.figure(figsize = (10, 7))
plt.scatter(CCTV_SDFP['인구수'], 
            CCTV_SDFP['소계'], 
            marker='.',
            s = 100,
            c = '#d62728')
plt.xlabel('인구수')
plt.ylabel('소계')
plt.show()
```

> ## 3) 회귀선 추가 - seaborn

```
import seaborn as sns

plt.figure(figsize = (12, 9))

sns.regplot(x = '인구수', y = '소계', data = CCTV_SDFP,
            line_kws = {'color':'red'})

for n in range(len(CCTV_SDFP)):
    plt.text(CCTV_SDFP['인구수'][n] * 1.02, 
             CCTV_SDFP['소계'][n] * 0.98, 
             CCTV_SDFP.index[n], 
             fontsize = 12)

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```
