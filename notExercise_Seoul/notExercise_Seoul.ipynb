# I. Data Preprocessing

```
import warnings
warnings.filterwarnings('ignore')
```

> ## 1) 'notExercise.xlsx' 파일 읽어오기

```
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/notExercise.xlsx'
DF = pd.read_excel(url)

DF.head()
```

* DataFrame 정보

```
DF.info()
```

> ## 2) '기간' 열(Column) 삭제

* '기간'열(Column) 정보

```
DF['기간'].value_counts()
```

* '기간'열(Column) <font color = 'red'>삭제

```
DF.drop('기간', axis = 1, inplace= True)
```

* <font color = 'red'>삭제</font> 결과 확인

```
DF.head()
```

> ## 3) 0, 23~52번(지역분류) index 삭제

* '대분류' 정보

```
DF['대분류']
```

* 23번 ~ 52번 index <font color = 'red'>삭제

```
DF.drop(DF.index[23:53], axis=0, inplace = True)
```

```
DF
```

* 0번 index <font color = 'red'>삭제

```
DF.drop(index=0, axis =0, inplace = True)
```

```
DF
```

# II. '성별' 통계

> ## 1) '성별' 데이터 추출

```
DF_sex = DF[DF['대분류'] == '성별']
```

```
DF_sex
```

> ## 2) '대분류' 열(Column) 삭제

```
DF_sex.drop('대분류', axis=1, inplace = True)
```

```
DF_sex
```

> ## 3) '분류'열(Column)을 index로 지정

```
DF_sex.set_index('분류', inplace = True)
```

```
DF_sex
```

> ## 4) matplotlib 시각화

```
fig, ax = plt.subplots(nrows =2, ncols =2, figsize=(15, 10))
ax[0,0].pie(DF_sex['운동을 할 충분한 시간이 없어서'],
        labels = DF_sex['운동을 할 충분한 시간이 없어서'].index,
        autopct = '%.1f%%',
        startangle= 90)
ax[0,1].pie(DF_sex['함께 운동을 할 사람이 없어서'],
        labels = DF_sex['함께 운동을 할 사람이 없어서'].index,
        autopct = '%.1f%%',
        startangle= 90)
ax[1,0].pie(DF_sex['운동을 할 만한 장소가 없어서'],
        labels = DF_sex['운동을 할 만한 장소가 없어서'].index,
        autopct = '%.1f%%',
        startangle= 90)
ax[1,1].pie(DF_sex['운동을 싫어해서'],
        labels = DF_sex['운동을 싫어해서'].index,
        autopct = '%.1f%%',
        startangle= 90)

ax[0,0].set_title('운동을 할 충분한 시간이 없어서')
ax[0,1].set_title('함께 운동을 할 사람이 없어서')
ax[1,0].set_title('운동을 할 만한 장소가 없어서')
ax[1,1].set_title('운동을 싫어해서')

plt.show()
```

# III. '연령별' 통계

> ## 1) '연령별' 데이터 전처리

```
DF_age = DF[DF['대분류'] == '연령별']
DF_age.drop('대분류', axis =1, inplace = True)
DF_age.set_index('분류', inplace = True)
```

```
DF_age
```

> ## 2) matplotlib 시각화

```
fig, ax = plt.subplots(ncols =3, figsize=(15, 10))
ax[0].pie(DF_age['운동을 할 충분한 시간이 없어서'],
        labels = DF_age['운동을 할 충분한 시간이 없어서'].index,
        autopct = '%.1f%%',
        startangle= 90)
ax[1].pie(DF_age['함께 운동을 할 사람이 없어서'],
        labels = DF_age['함께 운동을 할 사람이 없어서'].index,
        autopct = '%.1f%%',
        startangle= 90)
ax[2].pie(DF_age['운동을 할 만한 장소가 없어서'],
        labels = DF_age['운동을 할 만한 장소가 없어서'].index,
        autopct = '%.1f%%',
        startangle= 90)

ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[2].set_title('운동을 할 만한 장소가 없어서')

plt.show()
```
